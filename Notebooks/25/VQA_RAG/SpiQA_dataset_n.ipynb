```python
# Fix the lambda function
df['visual_type'] = df['reference_figure'].apply(
    lambda x: "Figure" if "figure" in x.lower() else ("Table" if "table" in x.lower() else "Other")
)
plt.figure()
df['visual_type'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',  # show percentages
    startangle=90,      # rotate so first slice starts at 12 o'clock
    colors=['sandybrown', 'indianred'],  # you can set colors manually
    wedgeprops={ 'linewidth': 1, 'alpha':0.6},
)
plt.ylabel('')  # Remove y-axis label
plt.title('Distribution of Visual Types')
plt.tight_layout()
plt.show()
```

#### 2.5.2. Visual componets comparison with other columns

```python
df_melted = df.melt(id_vars=['visual_type'], value_vars=['question_length_Vis', 'caption_length', 'answer_length'],
                    var_name='feature', value_name='length')
```

```python
palette = {'Figure': 'sandybrown', 'Table': 'indianred'}
sns.violinplot(
    data=df_melted,
    x='feature',
    y='length',
    hue='visual_type',
    palette=palette,
    split=True,
    inner='quartile',
    cut=0,
    bw_adjust=0.8
)
```

```python

fig, axes = plt.subplots(3, 1, figsize=(5, 12), sharey=False)
palette = {'Figure': 'sandybrown', 'Table': 'indianred'}

# Column names to plot
columns = ['question_length_Vis', 'caption_length', 'answer_length']
titles = ['Question Length by Visual Type', 'Caption Length by Visual Type', 'Answer Length by Visual Type']

for idx, (col, title) in enumerate(zip(columns, titles)):
    ax = axes[idx]
    
    sns.violinplot(
        data=df_melted,
        x='feature',
        y='length',
        hue='visual_type',
        palette=palette,
        split=True,
        inner='quartile',
        cut=0,
        bw_adjust=0.8
    )
    
    # Set alpha for each violin patch
    for patch in ax.collections:
        patch.set_alpha(0.6)

    # Set titles and labels nicely
    ax.set_title(title, fontsize=13)
    ax.set_xlabel('' if idx < 2 else 'Visual Type')  # Only last plot has x-axis label
    ax.set_ylabel('Number of Words')

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(3, 1, figsize=(5, 12), sharey=False)

palette = {'Figure': 'sandybrown', 'Table': 'indianred'}
columns = ['question_length_Vis', 'caption_length', 'answer_length']
titles = ['Question Length by Visual Type', 'Caption Length by Visual Type', 'Answer Length by Visual Type']

for idx, (col, title) in enumerate(zip(columns, titles)):
    ax = axes[idx]
    
    # Violin plot
    sns.violinplot(
        data=df,
        x='visual_type',
        y=col,
        hue='visual_type',
        palette=palette,
        split=True,
        inner='quartiles',
        linewidth=1.2,
        cut=0,
        bw_adjust=0.8,
        ax=ax,
        legend=False
    )

    # Set alpha for each violin
    for patch in ax.collections:
        patch.set_alpha(0.6)

    # Overlay tiny black dots (raw data points)
    sns.stripplot(
        data=df,
        x='visual_type',
        y=col,
        hue='visual_type',
        dodge=True,
        jitter=True,
        size=2,
        palette={'Figure': 'saddlebrown', 'Table': 'firebrick'},
        marker='o',
        ax=ax,
        legend=False
    )

    # Titles and labels
    ax.set_title(title, fontsize=13)
    ax.set_xlabel('' if idx < 2 else 'Visual Type')
    ax.set_ylabel('Number of Words')

plt.tight_layout()
plt.show()
```

## Get image paths

```python
import os

def get_image_paths(doc_id, question_idx, spiqa_d, image_root):
    """
    Get the full image path(s) for a given paper and question index from SPIQA.
    
    Parameters:
    - doc_id: e.g. "1506.06579v10"
    - question_idx: index of the question (e.g. 0, 1, 2)
    - spiqa_d: original SPIQA dataset as a dict
    - image_root: root directory where images are stored
    
    Returns:
    - List of full image paths
    """
    base_id = doc_id.split("v")[0]  # strip version
    paper = spiqa_d.get(base_id)

    if not paper:
        raise ValueError(f"No paper found for doc_id: {doc_id}")

    fig_keys = paper['referred_figures_tables'][question_idx]
    return [os.path.join(image_root, fig_key) for fig_key in fig_keys]
```
